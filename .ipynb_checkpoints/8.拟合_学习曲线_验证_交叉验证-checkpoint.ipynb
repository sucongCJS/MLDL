# 过拟合(overfitting) 欠拟合(underfitting)
# 学习曲线
> 随着训练样本的逐渐增多, 算法训练出的模型的表现能力
- 欠拟合学习曲线稳定的位置较高
- 过拟合学习曲线, 训练的误差小, 测试的误差大

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, size=100)
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def plot_learning_curve(algo, X_train, X_test, y_train, y_test):
    train_score = []
    test_score = []
    for i in range(1, len(X_train)+1):
        algo.fit(X_train[:i], y_train[:i])

        y_train_predict = algo.predict(X_train[:i])
        train_score.append(mean_squared_error(y_train[:i], y_train_predict))

        y_test_predict = algo.predict(X_test)
        test_score.append(mean_squared_error(y_test, y_test_predict))

    plt.plot(np.arange(1, len(X_train)+1), test_score, label="test")
    plt.plot(np.arange(1, len(X_train)+1), train_score, label="train")
    
    plt.legend()
    plt.axis([0, len(X_train)+1, 0, 4])
    plt.show()
    
lin_reg = LinearRegression()
plot_learning_curve(lin_reg, X_train, X_test, y_train, y_test)
```

### train的学习曲线:
- 误差随着样本的增加而增大, 因为样本越多越难拟合住全部的点

### test的学习曲线:
- 样本少的时候, 误差很大
- 测试误差比训练误差大, 因为是由训练模型泛化的

## 多项式回归的学习曲线
稳定时误差较小, 则说明泛化能力越强

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures 

def PolynomialRegression(degree):
    return Pipeline([
        ("poly", PolynomialFeatures(degree=degree)), # 第一个为名字
        ("std_scaler", StandardScaler()), 
        ("lin_reg", LinearRegression())
    ])
```

```python
ploy2_reg = PolynomialRegression(degree=2)
plot_learning_curve(ploy2_reg, X_train, X_test, y_train, y_test)
```

```python
# 过拟合
ploy2_reg = PolynomialRegression(degree=20)
plot_learning_curve(ploy2_reg, X_train, X_test, y_train, y_test)
```

# 验证数据集
train_test_split容易过分依赖于test, 也会导致针对测试数据集过拟合

- 训练数据: (参与模型的训练)
- 验证数据: 调整超参数使用的数据集 (参与模型的训练)
- 测试数据: 作为衡量最终模型性能的数据集 (不参与模型的训练)

# 交叉验证 k-folds
- 防止验证数据集中有极端的例子

## 先测试train_test_split

```python
import numpy as np
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=666)
```

```python
from sklearn.neighbors import KNeighborsClassifier

best_score, best_k, best_p = 0, 0, 0
for k in range(2, 5):
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(weights="distance", n_neighbors=k, p=p)
        knn_clf.fit(X_train, y_train)
        score = knn_clf.score(X_test, y_test)
        if score > best_score:
            best_score, best_k, best_p = score, k, p
            
print(best_k)
print(best_p)
print(best_score)
```

## 交叉验证

```python
from sklearn.model_selection import cross_val_score

knn_clf = KNeighborsClassifier()
cross_val_score(knn_clf, X_train, y_train, cv=3) # sklearn将数据分为3份
```

```python
best_score, best_k, best_p = 0, 0, 0
for k in range(2, 5):
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(weights="distance", n_neighbors=k, p=p)
        scores = cross_val_score(knn_clf, X_train, y_train, cv=3)
        score = np.mean(scores)
        if score > best_score:
            best_score, best_k, best_p = score, k, p
            
print(best_k)
print(best_p)
print(best_score)
```

```python
best_knn_clf = KNeighborsClassifier(weights="distance", n_neighbors=2, p=2)
```

```python
best_knn_clf.fit(X_train, y_train)
best_knn_clf.score(X_test, y_test) # 用从来没有用来训练模型的测试数据集来检测准确率
```
