## random

```python
np.random.randint(0, 10) # 左开右闭
```

```python
np.random.randint(4, 8, size=10)
```

```python
np.random.randint(4, 8, size=(3, 5))
```

```python
np.random.seed(666) # 用同个种子生成的随机数一样
np.random.randint(4, 8, size=(3, 5))
```

```python
np.random.randint(4, 8, size=(3, 5))
```

```python
np.random.random()
```

```python
import matplotlib.pyplot as plt
x = np.random.normal(1,10,2000) # 均值, 标准差, 大小
y = np.random.normal(1,10,2000)
plt.scatter(x,y, marker="+", alpha=0.5)
plt.show()
```

```python
# 均匀分布
x = np.random.uniform(1,2,2000)
y = np.random.uniform(1,2,2000)
plt.scatter(x,y, marker="+", alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = np.random.randn(2000) # 均值, 标准差, 大小
y = np.random.randn(2000)
plt.scatter(x,y, marker="+", alpha=0.5)
plt.show()
```
