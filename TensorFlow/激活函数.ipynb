# 激活函数
作用: 加入非线性

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
def sigmoid(inputs):
    return [1 / float(1 + np.exp(-x)) for x in inputs]

# 如果小于0, 则等于0
def relu(inputs):
    return [x * (x>0) for x in inputs]

def tanh(inputs):
    return [(np.exp(x) - np.exp(-x)) / float(np.exp(x) + np.exp(-x)) for x in inputs] # 分母为浮点, 结果也会自动变为浮点, 分子不需要float了

def softplus(inputs):
    return [np.log(1 + np.exp(x)) for x in inputs]
```

```python
x = np.linspace(-7, 7, 180)
y_sigmoid = tf.nn.sigmoid(x)
y_relu = tf.nn.relu(x)
y_tanh = tf.nn.tanh(x)
y_softplus = tf.nn.softplus(x)
with tf.Session() as sess:
    y_sigmoid, y_relu, y_tanh, y_softplus = sess.run([y_sigmoid, y_relu, y_tanh, y_softplus])
```

```python
plt.subplot(221)
plt.plot(x, y_sigmoid, label="Sigmoid")
plt.ylim(-0.2, 1.2)
plt.legend(loc="best")

plt.subplot(222)
plt.plot(x, y_relu, label="Relu")
plt.ylim(-1, 6)
plt.legend(loc="best")

plt.subplot(223)
plt.plot(x, y_tanh, label="Tanh")
plt.ylim(-1.3, 1.3)
plt.legend(loc="best")

plt.subplot(224)
plt.plot(x, y_softplus, label="softplus")
plt.ylim(-1, 6)
plt.legend(loc="best")
```
